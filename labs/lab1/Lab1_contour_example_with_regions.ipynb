```python
import numpy as np
import matplotlib.pyplot as plt

nbin=300
var_1 = np.linspace(-5,5,nbin)
var_2 = np.linspace(-5,5,nbin)
X,Y = np.meshgrid(var_1,var_2)
Z = X**2+Y**2


Z = np.clip(Z,0,20) #if you want

print(Z)

plt.figure(figsize=(8,6)) # make the 2-D contour plot
plt.rcParams['font.size'] = 12

plt.subplot(1,1,1) #1 subplot for contourf

plt.contourf(X,Y,Z,500,cmap='Blues_r') #make contourf
cbar = plt.colorbar()  # plot colorbar on right for contourf
levels = [2.30,6.17]  #from lecture 5, slide 3 for 2dof and 68%,95% regions
CS = plt.contour(X,Y,Z,levels,colors=['yellow','yellow'],linestyles=['solid','dashed'])
#next for labeling contours with text
fmt = {}
strs = ['68%', '95%']
for l, s in zip(CS.levels, strs):
    fmt[l] = s
plt.clabel(CS,CS.levels,inline=True,fmt=fmt) #put strings into labels
cbar.set_label(r'Z',fontsize=15)

plt.plot(0,0,'ro',label='0,0') #marker for best fit
plt.plot(1,1,'yP',label='1,1') #marker for True value
plt.xlabel(r'var_1',fontsize=15)
plt.ylabel(r'var_2',size=15)
plt.legend()
plt.title(r'Z vs var_1, var_2',size=18)



plt.tight_layout()
plt.show()
```
